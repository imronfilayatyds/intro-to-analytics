## Merge, Join, and Concatenate

```python
df1 = pd.read_csv("netflix_one.csv")
df1
```

```python
df2 = pd.read_csv("netflix_two.csv")
df2
```

```python
# d1.merge(d2, how = 'inner') --> intersection
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
# d1.merge(d2, how = 'outer') --> union
df1.merge(df2, how = 'outer')
```

```python
# d1.merge(d2, how = 'left') --> d1 + column from d2
df1.merge(df2, how = 'left')
```
